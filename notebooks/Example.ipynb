#### Inputs

Set the FFTLog, cosmological and nuisance parameters for which you want to compute the redshift space power spectrum multipoles. Upload the corresponding linear power spectrum. You can also compute it using an Einstein-Boltzmann code like CAMB, CLASS, etc.

```python
""" use: NeuEffect = 'y' to employ fk-kernels, EdS-kernels are used if NeuEffect = 'no'  """
NeuEffect = 'y'
```

```python
"FFTLog parameters"
kmin = 10**(-7); kmax = 100.; N = 128; 
biasnu = -0.1;   #Not yet tested for other values    
FFTLogParams = [kmin, kmax, N, biasnu]


"Cosmological parameters"
#omega_i = Omega_i hÂ² 
omega_b = 0.0220684;        #baryons
omega_cdm = 0.116631;       #CDM
omega_ncdm = 0.00429461;    #massive neutrinos 
h = 0.6711;                 #h = H0/100
A_s = 2.7391e-9;            #scalar amplitude
n_s = 0.9624;               #scalar spectral index 
z_pk = 0.5;                 #redshift evaluation
CosmoParams = [z_pk, omega_b, omega_cdm, omega_ncdm, h]


"Linear power spectrum"
inputpkT = np.loadtxt('pkl_cb_m04_z05.dat', unpack = True)


"Nuisance parameters"
b1 = 1.645;      
b2 = -0.46;      
bs2 = -4/7*(b1 - 1);        #coevolution     
b3nl = 32/315*(b1 - 1);     #coevolution
alpha0 = 3;  
alpha2 = -28.9;  
alpha4 = 0.0;   
ctilde = 0.0; 
alphashot0 = 0.08;
alphashot2 = -8.1;      
PshotP = 1/0.0002118763;
NuisanParams = [b1, b2, bs2, b3nl, alpha0, alpha2, alpha4, 
                ctilde, alphashot0, alphashot2, PshotP]
```

#### Computation of $M$ matrices 
This calculation is performed only one time. That is, the first time the code is called, it computes the $M$ matrices and stores them for the rest of the runs, which can be of the order of thousands in parameter estimations.

```python
#Cosmology-independent matrices
Matrices = M.Matrices(FFTLogParams)
```

#### Computation of the nonlinear modulus
Here the code computes all the 1-loop contributions through the FFTLog formalism for both the wiggle and non-wiggle (also computed by the code) linear power spectra. Finally the code computes the EFT galaxy redshift space power spectrum, includes IR resummations and calculates the Legendre multipoles.

```python
"Computing 1-loop corrections"
LoopCorrections = M.LoopCorr(inputpkT, CosmoParams, FFTLogParams, NeuEffect) 
    
"IR resummations and multipoles"
kh, Pkl0, Pkl2, Pkl4 = M.PklModel(inputpkT, CosmoParams, NuisanParams, 
                                      FFTLogParams, NeuEffect)
```

```python
fig, axs = plt.subplots(figsize=(7,5))
axs.set_xlabel(r'$k \, [h\, Mpc^{-1}]$', fontsize =  15)
axs.set_ylabel(r'$k P_{\ell}(k) \, [h^{-1} \,  Mpc]^2$', fontsize =  15)

axs.plot(kh, kh*Pkl0, color = 'navy', ls='-', label=r'$\ell = 0$')
axs.plot(kh, kh*Pkl2, color = 'maroon', ls='-', label=r'$\ell = 2$')
axs.plot(kh, kh*Pkl4, color = 'darkgreen', ls='-', label=r'$\ell = 4$')

axs.set_xlim([kh[0], 0.20])
axs.set_ylim([-200.0, 1800])

leg1 = axs.legend(loc = 'best');
leg1.get_frame().set_linewidth(0.0);

plt.show()
```
